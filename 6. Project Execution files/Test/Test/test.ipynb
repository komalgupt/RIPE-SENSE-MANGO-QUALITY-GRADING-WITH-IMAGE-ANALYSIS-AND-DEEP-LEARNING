```python
from keras.models import load_model
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from matplotlib import pyplot as plt
```

```python
loaded_model = load_model("Model testing and training file/vgg16_model.h5")
```

```python
test_image = load_img("Model testing and training file/mango1.jpeg", target_size = (128,128))
test_image_array = img_to_array(test_image)
```

```python
test_image_array = test_image_array/255.0
test_image_array = test_image_array.reshape((1,128,128,3))
prediction = loaded_model.predict(test_image_array)
#print the predicted class
predicted_class_index = np.argmax(prediction)
class_names = ['under_ripen','over_ripen','perfect_ripen']
predicted_class_name = class_names[predicted_class_index]
print("predicted class:",predicted_class_name)
#display the test image
plt.imshow(image.load_img("Model testing and training file/mango1.jpeg"))
plt.title(predicted_class_name)
plt.show()
```
